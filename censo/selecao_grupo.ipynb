```python
import pandas as pd
```

```python
teste = pd.read_csv('proppretorural.csv', sep = ';', encoding='utf-8')
print(teste)
```

```python
estados_nordeste = ["AL", "BA", "CE", "MA", "PB", "PE", "PI", "RN", "SE"]

teste['estado'] = teste['Município'].str[-3:-1]
eita = teste.head()
```

```python
base_filtrada = teste[teste['estado'].isin(estados_nordeste)]
print(base_filtrada)
```

```python
base_filtrada.drop(columns = ['Sexo','Idade','estado','Ano','Variável'], inplace = True)
```

```python
base_filtrada
```

```python
# iteração que pega a cidade depois especificamente rural preta e divide por total total.
dicio_geral = {}
dicio_pret = {}
for index, row in base_filtrada.iterrows():
    if row['Situação do domicílio'] == 'Total' and row["Cor ou raça"] == 'Total':
       dicio_geral[row['Município']] =  row['Quantidade']

    if row['Situação do domicílio'] == 'Rural' and row["Cor ou raça"] == 'Preta':
        dicio_pret[row['Município']] = row['Quantidade']
```

```python
dicio_pret['Açailândia (MA)']
```

```python
dicio_df = {}
for i in dicio_pret.keys():
        valor_pret = dicio_pret[i]
        valor_geral = dicio_geral[i]

        if valor_pret != '-' and valor_geral != '-':  # Verifica se os valores são válidos
            prop = int(valor_pret) / int(valor_geral)
            dicio_df[i] = prop

df = pd.DataFrame.from_dict(dicio_df, orient = 'index')

df.rename(columns= {0: 'Proporção rural/preto'}, inplace=True)
```

```python
# Renomear a coluna index para "Municipio" diretamente
df.rename_axis('Municipio', inplace=True)  
df.reset_index(inplace=True)
df.rename_axis(None, inplace=True)
```

```python
df_ordenado = df.sort_values("Proporção rural/preto").reset_index(drop=True)
print(df_ordenado.to_string())
```

```python
cidade_alvo = 'Santa Luzia do Itanhy (SE)'

indice = df_ordenado[df_ordenado["Municipio"] == cidade_alvo].index[0]

cidades_acima = df_ordenado.iloc[indice-8:indice]
cidades_abaixo = df_ordenado.iloc[indice+1:indice+9]

cidades = pd.concat([cidades_abaixo,cidades_acima])
```
