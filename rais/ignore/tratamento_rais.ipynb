```python
import pandas as pd
```

```python
# tentativa de leitura de uma das bases e tranformacao em um conjunto de csv
df = pd.read_csv('C:/Users/DESKTOP 15/Desktop/THP/rais/1011/RAIS_2010_anonimizado.csv', sep=";", encoding='latin-1', nrows=1000)
df.head()
```

```python
print(len(df.columns))
df.columns
```

```python
# Extrair o ano da coluna 'dtnascimento' e calcular a idade em anos
df['dtnascimento'] = pd.to_datetime(df['dtnascimento'], dayfirst=True)
ano_base = df['dtnascimento'].dt.year.max()  # Ano mais recente na base
df['Idade'] = ano_base - df['dtnascimento'].dt.year

# Calcular remuneração em salários mínimos (SM)
salario_minimo = 1200  # Valor hipotético do salário mínimo
df['Vl Remun Média (SM)'] = df['remmedia'] / salario_minimo
df['Vl Remun Média (SM) - ultrem'] = df['ultrem'] / salario_minimo

# Criar faixas etárias
bins = [0, 24, 34, 44, 54, float('inf')]
labels = ['Até 24 anos', '25 a 34 anos', '35 a 44 anos', '45 a 54 anos', '55 anos ou mais']
df['Faixa Etária'] = pd.cut(df['Idade'], bins=bins, labels=labels, right=False)

# Criação da coluna 'Vínculo Ativo 31/12'
df['Vínculo Ativo 31/12'] = df['tpvinculo'].apply(lambda x: 'Sim' if x == '10' else 'Não')

# Renomear colunas
df = df.rename(columns={'sbclas20': 'CNAE 2.0 Classe', 'municipio': 'Município', 'raca_cor': 'Raça Cor', 'genero': 'Sexo Trabalhador', 'tempempr': 'Tempo Emprego'})

# Selecionar colunas desejadas
df_final = df[['CNAE 2.0 Classe', 'Faixa Etária', 'Vl Remun Média (SM)', 'Vl Remun Média (SM) - ultrem', 'Idade', 'Município', 'Raça Cor', 'Sexo Trabalhador', 'Tempo Emprego', 'Vínculo Ativo 31/12']]

# Exibir as 5 primeiras linhas
print(df_final.head().to_markdown(index=False, numalign="left", stralign="left"))
```
